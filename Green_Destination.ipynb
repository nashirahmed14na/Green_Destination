```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gd= pd.read_csv('greendestination (1).csv')
```

```python
gd.head()
```

```python
gd.isnull().sum()
```

```python
gd.describe()
```

```python
gd["Attrition"]= gd["Attrition"].map({'Yes':1, 'No':0})
```

```python
gd.head(3)
```

```python
attrition_rate = gd['Attrition'].mean() * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(gd['Age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(gd['YearsAtCompany'], kde=True, bins=20)
plt.title('Years at Company Distribution')
plt.xlabel('Years at Company')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(gd['MonthlyIncome'], kde=True, bins=20)
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()
```
